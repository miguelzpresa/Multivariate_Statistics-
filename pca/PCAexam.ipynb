# PCA sin normalizar

```python
pca = PCA()
pca.fit(df_numericas)



explained_variance = pca.explained_variance_ratio_

# Varianza acumulada
cumulative_variance = np.cumsum(explained_variance)

# Graficar la varianza explicada y la varianza acumulada
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(explained_variance) + 1), cumulative_variance, marker='o', linestyle='--')
plt.title('Varianza Explicada Acumulada por el PCA')
plt.xlabel('Número de Componentes')
plt.ylabel('Varianza Explicada Acumulada')
plt.grid(True)
plt.show()
```

```python
pca = PCA( n_components=2 )
compo2 = pca.fit_transform(df_numericas)

print(compo2.shape)
df_pca = pd.DataFrame(data=compo2, columns=["PC1", "PC2"])
df_pca
```

# PCA con xi/max(abs(xi))

```python
abs = df_numericas.abs()
scaled =  df_numericas/abs.max()

pca = PCA()
pca.fit(scaled)

explained_variance = pca.explained_variance_ratio_

# Varianza acumulada
cumulative_variance = np.cumsum(explained_variance)

# Graficar la varianza explicada y la varianza acumulada
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(explained_variance) + 1), cumulative_variance, marker='o', linestyle='--')
plt.title('Varianza Explicada Acumulada por el PCA')
plt.xlabel('Número de Componentes')
plt.ylabel('Varianza Explicada Acumulada')
plt.grid(True)
plt.show()
```

```python
pca = PCA( n_components=3 )
compo2 = pca.fit_transform(scaled)

print(compo2.shape)
df_pca = pd.DataFrame(data=compo2, columns=["PC1", "PC2","PC3"])
df_pca
```

#  PCA con e^^xi

```python
exponential_df = df_numericas.apply(lambda x: np.exp(x))
exponential_df.replace([np.inf, -np.inf], -1, inplace=True)
#exponential_df = exponential_df[~(exponential_df == -1).any(axis=1)]

pca = PCA()
pca.fit(exponential_df)
explained_variance = pca.explained_variance_ratio_

# Varianza acumulada
cumulative_variance = np.cumsum(explained_variance)

# Graficar la varianza explicada y la varianza acumulada
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(explained_variance) + 1), cumulative_variance, marker='o', linestyle='--')
plt.title('Varianza Explicada Acumulada por el PCA')
plt.xlabel('Número de Componentes')
plt.ylabel('Varianza Explicada Acumulada')
plt.grid(True)
plt.show()
```

```python
pca = PCA( n_components=3 )
compo2 = pca.fit_transform(combi_lin)

print(compo2.shape)
df_pca = pd.DataFrame(data=compo2, columns=["PC1", "PC2","PC3"])
df_pca
```

### calculamos los coeficientes Beta para el dataset original

```python
y = df_numericas["price"]
df_numericas = df_numericas.drop(columns=["price"])
xtrain, xtest, ytrain, ytest = train_test_split(df_numericas, y, train_size=0.8)
lir = LinearRegression()

# Ajustar el modelo
lir.fit(xtrain, ytrain)

# Predecir los valores
ypred = lir.predict(xtest)

# Mostrar los resultados
print("Coeficientes:", lir.coef_)
print("Intercepto:", lir.intercept_)
print("R^2 en el conjunto de entrenamiento:", lir.score(xtrain, ytrain))
print("R^2 en el conjunto de prueba:", lir.score(xtest, ytest))
```
